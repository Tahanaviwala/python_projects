## Setup

The following code imports the required libraries and loads the ```grades.csv``` file into a Pandas DataFrame.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
````

````python
def normal_distribution(x, mu, sigma):
    return 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))
````

````python
df = pd.read_csv('grades.csv')
````

## Density of the final grades

Make a plot that meets the following criteria:
* normalized histogram with 20 bins in the range 0 to 100 using the Final Grade column
* a line plot of the normal distribution using the mean and standard deviation of the final grades
* appropriate labels on the both the x and y axes
* appropriate title
* appropriate legend

````python
x = df['Final Grade']
x
````

````python
from statistics import mean,stdev
mu = mean(x)
sigma = stdev(x)
sigma
y = normal_distribution(x,mu,sigma)
````

````python
bins = 20
plt.title("Grade Distribution")
plt.xlabel("Numeric Grade")
plt.ylabel("Normalized Frequency")
labels = ["Normal Distribution", "Probability Mass"]
x_min = 0
x_max = 100
value = np.linspace(0, 100, 100)
y = normal_distribution(value, mu, sigma)
plt.plot(value,y, color='coral')
plt.hist(x, bins, range=[0,100] ,edgecolor='black', density=True)
plt.legend(labels,loc="upper left")
plt.show()
````

## Bar chart of final letter grades

Make a plot the meets the following criteria:
* horizontal bar chart of the final letter grades
* ten evenly spaced tick marks on the x axis from 0 to the maximum mean price
* appropriate labels on the x and y axes
* appropriate title

The letter grade scale is as follows:
* A: 90 - 100
* B: 80 - 89
* C: 70 - 79
* D: 60 - 69
* F: 0 - 59

````python
Data = np.array(df['Final Grade'])
Data = np.sort(Data, axis=0)
score = [0.0, 60.0, 70.0, 80.0, 90.0, 100.0]
grades = {'A':0, 'B':0, 'C':0, 'D':0,  'F':0}


for j in range(len(Data)):
    if (Data[j] >= 90 and Data[j] < 100 ):
            grades['A']  +=1
    elif(Data[j] >= 80 and Data[j] < 90 ):        
            grades['B']  +=1
    elif(Data[j] >= 70 and Data[j] < 80 ):        
            grades['C']  +=1
    elif(Data[j] >= 60 and Data[j] < 70 ):        
            grades['D']  +=1
    else:      
            grades['F']  +=1

    
# grades {'A': 6, 'B': 23, 'C': 21, 'D': 21, 'F': 29}  results in dictionary

y_pos = np.arange(len(grades.keys()))
plt.barh(y_pos, grades.values(), align='center')
plt.yticks(y_pos, grades.values())
plt.xlabel('No of Students')
plt.title('Lette Grade')
plt.show()
````

````python
grades  = dict(sorted(grades.items(), key=lambda x: x[1] , reverse = True))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.barh(list(grades.keys()), list(grades.values()), align='center', height=0.5, edgecolor="black")
ax.set_yticks(list(grades.keys()))
ax.set_xlabel('Number of students')
ax.set_ylabel('Letter Grades')

ax.set_title('Letter Grades Count')
plt.tight_layout()
plt.show()
````
